```python
from joblib import Parallel, delayed
import lightgbm as lgb
import optuna
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

def generar_caracteristicas(df_input, historical_stats=None):
    df = df_input.copy()

    df['day'] = df['KeyWeek'].dt.day
    df['week_of_month'] = np.ceil(df['day'] / 7).astype(int)
    df['week_of_year'] = df['KeyWeek'].dt.isocalendar().week.astype(int)
    df['month'] = df['KeyWeek'].dt.month
    df['year'] = df['KeyWeek'].dt.year  # nuevo

    if historical_stats is not None:
        df['avg_demand_week_month'] = df['week_of_month'].map(historical_stats['avg_demand_week_month'])
        df['avg_demand_week_year'] = df['week_of_year'].map(historical_stats['avg_demand_week_year'])
        df['avg_sales_month'] = df['month'].map(historical_stats['avg_sales_month'])
        df['avg_sales_last4'] = historical_stats['avg_sales_last4']
    else:
        df['avg_demand_week_month'] = df.groupby(['KeyProduct', 'week_of_month'])['Demand'].transform('mean')
        df['avg_demand_week_year'] = df.groupby(['KeyProduct', 'week_of_year'])['Demand'].transform('mean')
        df['avg_sales_month'] = df.groupby(['KeyProduct', 'month'])['Demand'].transform('mean')
        df['avg_sales_last4'] = df.groupby('KeyProduct')['Demand'].transform(lambda x: x.rolling(window=4, min_periods=1).mean())

    df.drop(columns=['day'], inplace=True)
    return df

def objective(trial, ts_df):
    ts_df_feat = generar_caracteristicas(ts_df)

    # Separar el 칰ltimo punto para la validaci칩n
    X = ts_df_feat[['week_of_month', 'avg_demand_week_month', 'week_of_year',
                    'avg_demand_week_year', 'month', 'avg_sales_month', 'avg_sales_last4', 'year']]
    y = ts_df_feat['Demand']

    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=1, shuffle=False)

    params = {
        #'objective': trial.suggest_categorical('objective', ['regression']),# 'poisson', 'tweedie']),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.1),
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'num_leaves': trial.suggest_int('num_leaves', 8, 512),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 1.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 1.0),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.1, 1.0),
        #'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0)
    }

    model = lgb.LGBMRegressor(**params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    mse = mean_squared_error(y_val, y_pred)

    return mse

def forecast_lgbm(ts_df, forecast_weeks):
    ts_df_feat = generar_caracteristicas(ts_df)

    X_train = ts_df_feat[['week_of_month', 'avg_demand_week_month', 'week_of_year',
                          'avg_demand_week_year', 'month', 'avg_sales_month', 'avg_sales_last4', 'year']]
    y_train = ts_df_feat['Demand']

    study = optuna.create_study(direction='minimize')
    study.optimize(lambda trial: objective(trial, ts_df), n_trials=3)

    best_params = study.best_params
    model = lgb.LGBMRegressor(**best_params)
    model.fit(X_train, y_train)

    historical_stats = {
        'avg_demand_week_month': ts_df_feat.groupby('week_of_month')['Demand'].mean().to_dict(),
        'avg_demand_week_year': ts_df_feat.groupby('week_of_year')['Demand'].mean().to_dict(),
        'avg_sales_month': ts_df_feat.groupby('month')['Demand'].mean().to_dict(),
        'avg_sales_last4': ts_df_feat['avg_sales_last4'].iloc[-1]
    }

    future_df = pd.DataFrame({'KeyWeek': forecast_weeks})
    future_df['KeyProduct'] = ts_df['KeyProduct'].iloc[0]
    future_df = generar_caracteristicas(future_df, historical_stats)

    X_future = future_df[['week_of_month', 'avg_demand_week_month', 'week_of_year',
                          'avg_demand_week_year', 'month', 'avg_sales_month', 'avg_sales_last4', 'year']]

    future_df['Forecast'] = model.predict(X_future)

    return future_df[['KeyWeek', 'Forecast']]
#def generar_caracteristicas(df_input, historical_stats=None):
#    df = df_input.copy()
#
#    df['day'] = df['KeyWeek'].dt.day
#    df['week_of_month'] = np.ceil(df['day'] / 7).astype(int)
#    df['week_of_year'] = df['KeyWeek'].dt.isocalendar().week.astype(int)
#    df['month'] = df['KeyWeek'].dt.month
#
#    if historical_stats is not None:
#
#        df['avg_demand_week_month'] = df['week_of_month'].map(historical_stats['avg_demand_week_month'])
#        df['avg_demand_week_year'] = df['week_of_year'].map(historical_stats['avg_demand_week_year'])
#        df['avg_sales_month'] = df['month'].map(historical_stats['avg_sales_month'])
#        df['avg_sales_last4'] = historical_stats['avg_sales_last4']
#    else:
#
#        df['avg_demand_week_month'] = df.groupby(['KeyProduct', 'week_of_month'])['Demand'].transform('mean')
#        df['avg_demand_week_year'] = df.groupby(['KeyProduct', 'week_of_year'])['Demand'].transform('mean')
#        df['avg_sales_month'] = df.groupby(['KeyProduct', 'month'])['Demand'].transform('mean')
#        df['avg_sales_last4'] = df.groupby('KeyProduct')['Demand'].transform(lambda x: x.rolling(window=4, min_periods=1).mean())
#
#    df.drop(columns=['day'], inplace=True)
#    return df
#
#
#def forecast_lgbm(ts_df, forecast_weeks):
#    ts_df_feat = generar_caracteristicas(ts_df)
#
#    X_train = ts_df_feat[['week_of_month', 'avg_demand_week_month', 'week_of_year',
#                          'avg_demand_week_year', 'month', 'avg_sales_month', 'avg_sales_last4']]
#    y_train = ts_df_feat['Demand']
#
#    model = lgb.LGBMRegressor(n_estimators=100)
#    model.fit(X_train, y_train)
#
#    historical_stats = {
#        'avg_demand_week_month': ts_df_feat.groupby('week_of_month')['Demand'].mean().to_dict(),
#        'avg_demand_week_year': ts_df_feat.groupby('week_of_year')['Demand'].mean().to_dict(),
#        'avg_sales_month': ts_df_feat.groupby('month')['Demand'].mean().to_dict(),
#        'avg_sales_last4': ts_df_feat['avg_sales_last4'].iloc[-1]  
#    }
#
#    future_df = pd.DataFrame({'KeyWeek': forecast_weeks})
#    future_df['KeyProduct'] = ts_df['KeyProduct'].iloc[0]
#    future_df = generar_caracteristicas(future_df, historical_stats)
#
#    X_future = future_df[['week_of_month', 'avg_demand_week_month', 'week_of_year', 
#                          'avg_demand_week_year', 'month', 'avg_sales_month', 'avg_sales_last4']]
#
#    future_df['Forecast'] = model.predict(X_future)
#    
#    return future_df[['KeyWeek', 'Forecast']]
#
def forecast_arima(ts_df, forecast_weeks):
    ts_df = ts_df.sort_values('KeyWeek')
    series = ts_df.set_index('KeyWeek')['Demand']

    model = ARIMA(series, order=(2,1,1))
    model_fit = model.fit()

    forecast = model_fit.forecast(steps=len(forecast_weeks))

    forecast_df = pd.DataFrame({'KeyWeek': forecast_weeks, 'Forecast': forecast.values})
    
    return forecast_df

def forecast_simple(ts_df, forecast_weeks):
    promedio_general = ts_df['Demand'].mean()
    ultimo_valor = ts_df.sort_values('KeyWeek')['Demand'].iloc[-1]
    
    ts_df_feat = generar_caracteristicas(ts_df)
    avg_weekmonth = ts_df_feat.groupby('week_of_month')['Demand'].mean().to_dict()
    
    forecast_list = []
    for dt in forecast_weeks:
        week_of_month = int(np.ceil(dt.day / 7))
        prom_semana_mes = avg_weekmonth.get(week_of_month, promedio_general) 
        forecast_value = (prom_semana_mes + promedio_general + ultimo_valor) / 3
        forecast_list.append(forecast_value)
    
    forecast_df = pd.DataFrame({'KeyWeek': forecast_weeks, 'Forecast': forecast_list})
    return forecast_df

def process_group(ks, group, forecast_weeks):
    try:
        group = group.sort_values('KeyWeek')
        length = len(group)

        if length > 52:
            forecast_df = forecast_lgbm(group, forecast_weeks)
        elif length > 25:
            forecast_df = forecast_arima(group, forecast_weeks)
        else:
            forecast_df = forecast_simple(group, forecast_weeks)
            
        forecast_df['KeyProduct'] = ks
        return forecast_df
    except Exception as e:
        print(f"Error al procesar KeyProduct {ks}: {e}")
        return None

forecast_weeks = pd.date_range(start='2025-03-17', periods=25, freq='W-MON')

result_list = Parallel(n_jobs=-1)(
    delayed(process_group)(ks, group, forecast_weeks) for ks, group in df.groupby('KeyProduct')
)

result_list = [res for res in result_list if res is not None]

result = pd.concat(result_list, axis=0, ignore_index=True)
result = result[['KeyWeek', 'KeyProduct', 'Forecast']]

# Guardar resultados
output_file = "/home/ec2-user/celes-ml/tottus/result/result_tottus3.csv"
result.to_csv(output_file, index=False)
print(f"Archivo exportado a: {output_file}")#158min
```
